* Create dataset directory and subfolders

* Categorized disease image lists

```python
my_fung = arr.tolist(disease_mapped_pics[0]["mycosis-fungoides"])
squ_cell = arr.tolist(disease_mapped_pics[1]["squamous-cell-carcinoma-in-situ"])
basal = arr.tolist(disease_mapped_pics[2]["basal-cell-carcinoma"])
mel_nev = arr.tolist(disease_mapped_pics[3]["melanocytic-nevi"])
ver_vul = arr.tolist(disease_mapped_pics[4]["verruca-vulgaris"])
epi_cys = arr.tolist(disease_mapped_pics[5]["epidermal-cyst"])
seb_ker = arr.tolist(disease_mapped_pics[6]["seborrheic-keratosis"])
```

```python
def move_images(img_path, folder):
    for i in range(len(img_path)):
        # Define source and destination paths
        source_file = "ddidiversedermatologyimages\\"+ img_path[i]
        destination_folder = "dataset\\" + folder

        try:
            # Move the file
            shutil.move(source_file, destination_folder)
            print(f"'{os.path.basename(source_file)}' moved successfully to '{destination_folder}'")
        except FileNotFoundError:
            print(f"Error: Source file '{source_file}' not found.")
        except Exception as e:
            print(f"An error occurred: {e}")
```

* Move images into respective subfolders

```python
move_images(my_fung,"mycosis-fungoides")
```

```python
move_images(squ_cell,"squamous-cell-carcinoma-in-situ")
move_images(basal,"basal-cell-carcinoma")
move_images(mel_nev,"melanocytic-nevi")
move_images(ver_vul,"verruca-vulgaris")
move_images(epi_cys,"epidermal-cyst")
move_images(seb_ker,"seborrheic-keratosis")
```
